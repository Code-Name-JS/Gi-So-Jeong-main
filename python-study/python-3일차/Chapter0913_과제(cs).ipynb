[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요

얕은 복사는 객체의 최상위 레벨만 복사한다. 그래서 중첩된 객체의 변경 사항은 복사본과 원본에 모두 영향을 미친다!

깊은 복사는 객체를 재귀적으로 완전히 복사한다. 그래서 원본 객체를 수정해도 복사본에 영향을 미치지 않는다!
