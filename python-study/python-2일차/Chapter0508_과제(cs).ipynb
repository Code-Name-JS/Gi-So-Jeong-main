[문제1]

Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요


### Python의 변수 할당은 **참조(referenc) 개념**을 사용한다.
즉, 변수는 객체의 값을 직접 저장하는 것이 아니라, 객체를 가리키는 **포인터** 역할을 한다.

---

- C : 변수는 메모리 주소에 값을 직접 저장한다.   

    - 기본적으로 **값 복사**가 일어난다.

- Java : 객체에 대한 변수는 Python과 비슷하게 참조를 저장한다.

    - 기본 자료형(int, char 등)은 값 복사가 일어난다.

- JavaScript : 객체는 Python과 유사하게 참조로 할당된다.

    - 원시 값(숫자, 문자열 등)은 값 복사가 된다.
